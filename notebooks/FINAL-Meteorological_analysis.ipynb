```python

colors = ['#d98088','#e6c095','#9bb7e0','#b9e69c']
my_palette = sns.color_palette(colors)

g = sns.FacetGrid(data=df.loc[df['flag_full']==True], col='Season', col_wrap=4, hue='Season', 
                  palette=my_palette, height=3, aspect=1.3)

g.map_dataframe(sns.lineplot, x='Hour', y='u*', style='Month_n',
                linewidth=3, zorder=5)

g.refline(y=0.2) # Friction velocity threshold

for ax, c in zip(g.axes.ravel(), colors):
    h, l = ax.get_legend_handles_labels()
    for i in h:
        i.set_color(c)    
    ax.legend(h[3:6], l[3:6], loc='upper left')
    
for season, ax in g.axes_dict.items():
    ax.text(0.75, 0.85, season, transform=ax.transAxes, fontweight='bold')
    
    sns.lineplot(data=df.loc[df['Season']!=season].groupby(['Hour', 'Month']).mean(),
                 x='Hour', y='u*', 
                 units='Month',estimator=None,  color='.7', linewidth=1, ax=ax, legend=False)
    
ax.set_xticks(np.arange(0, 24, 4))
ax.set_xlim(0,23)
g.set_titles("")
g.set_axis_labels("Hour", "$u*$ $\\left[m \: s^{-1}\\right]$")
g.tight_layout()

# g.savefig(imgFolder/'meteorological-frictionvel_v2.png', dpi=300)
```
